#ตัดเกรด

```python
grades = {
    80: "A",
    75: "B+",
    70: "B",
    65: "C+",
    60: "C",
    55: "D+",
    50: "D",
    0: "F",
}

def grade_mapping(value):
    for key, letter in grades.items():
        if value >= key:
            return letter
```

```python
letter_grades = Allscore["คะแนนรวม"].map(grade_mapping)
Allscore["เกรด"] = pd.Categorical(
    letter_grades, categories=grades.values(), ordered=True)
Allscore
```
