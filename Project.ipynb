# ราคาบ้านเดี่ยวในกรุงเทพและปริมณฑล พร้อมกับคะแนนอาหารการกินของย่านต่างๆ และคะแนนการเดินทางเข้าถึงสถานที่ต่างๆ ในชีวิตประจำวัน เพื่อประกอบการตัดสินใจซื้อบ้านเดี่ยว

## ข้อมูลที่อยู่อาศัยราย Unit

```python
open_unit = pd.read_csv('/content/drive/MyDrive/Project Dataviz/Project Dataviz/opendata_unittype/opendata_unittype.csv')
open_unit.head()
```

```python
#Select columns
open_unit.columns
```

```python
unit = open_unit[['unittype_id', 'project_id','propertytype_id','propertytype_name_en', 'propertytype_name_th', 'area_total_min_wa','price_min']]
```

```python
#เลือกเฉพาะบ้าน
unit_BKK = unit[unit['propertytype_name_en']=='Detached House']
```

```python
#เช็ค missing ของ unit
unit_BKK.isnull().any()
```

```python
unit_BKK['area_total_min_wa'] = unit_BKK['area_total_min_wa'].fillna('1')
unit_BKK['price_min'] = unit_BKK['price_min'].fillna('0')
```

```python
unit_BKK.isnull().any() #เช็ค missing อีกครั้ง
```

```python
unit_BKK
```

## คะแนนเกี่ยวกับอาหารการกินของย่าน (Subdistrict) ต่างๆ

```python
open_eat = pd.read_csv('/content/drive/MyDrive/Project Dataviz/Project Dataviz/opendata_eating_score/opendata_eating_score.csv')
open_eat.head()
```

```python
#Select columns
open_eat.columns
```

```python
eat = open_eat[['district_id','province_name_th','eating_daytime', 'eating_daytime_price','eating_daytime_quality', 'eating_daytime_variety', 'eating_nighttime','eating_nighttime_price', 'eating_nighttime_quality','eating_nighttime_variety']]
```

##คะแนนการเดินทางเข้าถึงสถานที่ ในชีวิตประจำวันของพื้นที่ต่างๆ โดยคำนวณจาก 3 รูปแบบการเดินทาง และ 5 ประเภทของสถานที่เป้าหมายเพื่อการดำเนินชีวิตในแต่ละวัน
### คะแนนการเดินทางเข้าถึงสถานที่ คือ ความสะดวกต่างๆ เช่น ทางเดินเท้าดี หลอดไฟเยอะ สว่าง เป็นต้น
### 3 รูปแบบการเดินทาง : เดิน, ใช้ขนส่งสาธารณะ และรถส่วนตัว### 
### คะแนนสถานที่ คือ ความสะดวก ความใกล้ที่ต่างๆ เช่น อยู่ใกล้ รร ที่ทำงานเป็นต้น
### 5 ประเภทของสถานที่ : พักผ่อนหย่อนใจ, สถานที่บริการสาธารณะ, สถานศึกษา, shopping และแหล่งทำงาน

```python
open_living = pd.read_csv('/content/drive/MyDrive/Project Dataviz/Project Dataviz/opendata_living_score/opendata_living_score.csv')
open_living.head()
```

```python
#Select columns
open_living.columns
```

```python
living = open_living[['district_id','province_name_th','walk_leisure','walk_public_service','walk_school', 'walk_shopping', 'walk_work', 'walk_total','access_boat', 'access_bus', 'access_train', 'access_total','drive_leisure', 'drive_public_service', 'drive_school','drive_shopping', 'drive_work', 'drive_total']]
```

```python
#เช็ค missing ของ living
living.isnull().any()
```

```python
#เลือกเฉพาะกรุงเทพ
living_BKK = living[living['province_name_th']=='กรุงเทพมหานคร']
```

```python
living_BKK
```

## Merge ตารางเข้าด้วยกัน

```python
unit_project = unit_BKK.merge(province_BKK, left_on='project_id', right_on='project_id')
unit_project
```

```python
unit_project_eat = unit_project.merge(eat_BKK, left_on='district_id' , right_on='district_id')
unit_project_eat
```

```python
unit_project_eat_living = unit_project_eat.merge(living_BKK, left_on='district_id' , right_on='district_id')
unit_project_eat_living
```

```python
unit_project_eat_living.isnull().any() #เช็ค missing
```

```python
#Select columns
unit_project_eat_living.columns
```

```python
u_p_e_l = unit_project_eat_living[['unittype_id','project_id','propertytype_id','propertytype_name_th','area_total_min_wa','price_min','name_th','subdistrict_id','subdistrict_name_th', 'district_id','district_name_th', 'province_id', 'province_name_th', 'eating_daytime', 'eating_daytime_price',
       'eating_daytime_quality', 'eating_daytime_variety', 'eating_nighttime',
       'eating_nighttime_price', 'eating_nighttime_quality',
       'eating_nighttime_variety', 'province_name_th', 'walk_leisure',
       'walk_public_service', 'walk_school', 'walk_shopping', 'walk_work',
       'walk_total', 'access_boat', 'access_bus', 'access_train',
       'access_total', 'drive_leisure', 'drive_public_service', 'drive_school',
       'drive_shopping', 'drive_work', 'drive_total']]
```

```python
u_p_e_l.isnull().any() #เช็ค missing
```
